## Reorganizing Arrays

```python
before = np.array([[1,2,3,4],[5,6,7,8]])

before.shape
```

```python
after = before.reshape(8,1)
print(after)
```

```python
after = before.reshape(2,2,2)
print(after)
```

```python
# Vertically Stacking Vectors

v1 = np.array([1,2,3,4])

v2 = np.array([5,6,7,8])

np.vstack([v1,v2,v2,v2])
```

```python
# Horizontal Stack

h1 = np.ones((2,4))

h2 = np.zeros((2,2))

np.hstack((h1,h2))
```
